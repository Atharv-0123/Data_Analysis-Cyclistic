```python
import duckdb
# Convert timedelta to seconds
df['ride_length_seconds'] = df['ride_length'].dt.total_seconds()

# SQL query with DuckDB
query = """
SELECT member_casual, AVG(ride_length_seconds) AS avg_ride_length_seconds
FROM df
GROUP BY member_casual
"""

# Execute the query
result = duckdb.query(query).df()
# Convert seconds to minutes and seconds
#result['avg_ride_length_minutes'] = (result['avg_ride_length_seconds'] // 60).astype(int)
#result['avg_ride_length_secs'] = (result['avg_ride_length_seconds'] % 60).round(2)

# Drop unnecessary columns if needed
#result = result[['member_casual', 'avg_ride_length_minutes', 'avg_ride_length_secs']]

#print(result)

def convert_seconds_to_minutes_and_format(df, column_name):
    """
    Converts a column of seconds to a formatted string of 'X minutes Y secs'.
    
    Args:
    - df: The DataFrame containing the column to convert.
    - column_name: The name of the column that contains the time in seconds.
    
    Returns:
    - The DataFrame with an additional column containing the formatted time.
    """
    # Convert seconds to minutes and seconds
    df['minutes'] = (df[column_name] // 60).astype(int)
    df['secs'] = (df[column_name] % 60).round(2)
    
    # Combine minutes and seconds into a single formatted string
    df['formatted_time'] = df['minutes'].astype(str) + " minutes " + df['secs'].astype(str) + " secs"
    
    # Drop intermediate columns ('minutes' and 'secs')
    df.drop(['minutes', 'secs'], axis=1, inplace=True)
    
    return df
    
result = convert_seconds_to_minutes_and_format(result, 'avg_ride_length_seconds')
print(result)
```

```python
query = "SELECT member_casual,day_of_week,AVG(ride_length_seconds) AS avg_ride_length_secs FROM df GROUP BY member_casual, day_of_week ORDER BY member_casual, day_of_week"

result = duckdb.query(query).df()

result = convert_seconds_to_minutes_and_format(result, 'avg_ride_length_secs')

print(result)
```
